### Dummy Variable Creation

```python
df = pd.get_dummies(df, columns = ["3D_available","Genre"], drop_first = True)
```

### X-y Split

```python
y = df['Collection']
```

```python
y.head()
```

```python
X = df.loc[:, df.columns!='Collection']
```

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

### Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

### Standardizing Data

* we will convert the mean & variance of each of the variable to be 0 and 1.

* Since in SVN we are calculating distance and distance depends upon the scale of each variable therefore standardiztion is reqd.
* Otherwise, the variable with larger scale will have more impact on model than the one with smaller scale.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
sc
```

```python
X_train_std = sc.transform(X_train)
```

```python
X_train_std
```

```python
X_test_std = sc.transform(X_test)
```

```python
X_test_std #after standardization
```

```python
X_test #before Standardiztion
```

### Training Regression Tree

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel = 'linear', C = 500)
```

```python
svr.fit(X_train_std, y_train)
```

### Predict values using Trained Model

```python
y_test_pred = svr.predict(X_test_std)
y_train_pred = svr.predict(X_train_std)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```
